```python
std1=[95,39,54,83,92]
std2=[94,83,23,94,85]
marks=["Eng","Mat","Phy","Chem","Cs"]
plt.subplot(2,1,1) #Syntax(rows,columns,position) (2,1,1) = graph split into 2 rows and 1 columns. 1st position
plt.plot(marks,std1,label='student1',color='m')
plt.legend()
plt.subplot(2,1,2) #(2,1,1) = graph split into 2 rows and 1 columns. 2st position
plt.plot(marks,std2,label='student2',color='y')
plt.legend()
```

```python
std1=[95,39,54,83,92]
std2=[94,83,23,94,85]
std3=[44,22,64,13,77]
std4=[77,44,24,17,53]
marks=["Eng","Mat","Phy","Chem","Cs"]
plt.subplot(2,2,1) #Syntax(rows,columns,position) (2,2,1) = graph split into 2 rows and 2 columns. 1st position
plt.plot(marks,std1,label='student1',color='m')
plt.legend()
plt.subplot(2,2,2) #(2,2,1) = graph split into 2 rows and 2 columns. 2st position
plt.plot(marks,std2,label='student2',color='y')
plt.legend()
plt.subplot(2,2,3) #Syntax(rows,columns,position)
plt.plot(marks,std3,label='student3',color='r')
plt.legend()
plt.subplot(2,2,4) #Syntax(rows,columns,position)
plt.plot(marks,std4,label='student4',color='c')
plt.legend()
```

```python
RevA=[230,560,780,127,128]
RevB=[200,160,270,127,400]
profA=np.diff(RevA)
print("profit of A",profA)
profB=np.diff(RevB)
print("profit of B",profB)
X=['19-20','20-21','21-22','22-23']
plt.subplot(1,2,1)
plt.bar(X,profA,color='k',label='Profit A')
plt.legend(loc='best') #best location where you can put the legend
plt.subplot(1,2,2)
plt.bar(X,profB,color='g',label='Profit B')
plt.legend(loc='best')
```
